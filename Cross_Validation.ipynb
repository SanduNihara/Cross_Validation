```python
import pandas as pd
data = pd.read_csv("/content/drive/MyDrive/ML Evaluation/Iris.csv")
data = data.drop("Id",axis=1)
```

```python
x = data.drop("Species",axis=1)
x.head(5)
```

```python
y = data["Species"]
y.head(5)
```

## **Method 1 : Randomized Data spliting into test,train and testing accuracy**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

# K-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(x_train,y_train)
knn.score(x_test,y_test)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
rf.score(x_test,y_test)
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train, y_train)
nb.score(x_test,y_test)
```

# Support Vector Machine

```python
from sklearn.svm import SVC
svm = SVC()
svm.fit(x_train, y_train)
svm.score(x_test,y_test)
```

# **Method 2 : Using Cross validation, data spliting and testing accuracy**

```python
from sklearn.model_selection import cross_val_score
```

# K-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
cross_val_score(knn,x,y,cv=5).mean()
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
cross_val_score(rf,x,y,cv=5).mean()
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
cross_val_score(nb,x,y,cv=5).mean()
```

# Support Vector Machine

```python
from sklearn.svm import SVC
svm = SVC()
cross_val_score(svm,x,y,cv=5).mean()
```
